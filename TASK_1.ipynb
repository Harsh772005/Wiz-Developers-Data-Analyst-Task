```python
import numpy as  np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(r'Upi_app_transection_2021.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Calculate total transaction volume
total_volume = df['Volume (Mn)'].sum()

# Group data by bank/payment app and calculate total transaction volume
grouped_data = df.groupby('UPI Banks')['Volume (Mn)'].sum().reset_index()

# Sort the data by transaction volume in descending order
sorted_data = grouped_data.sort_values(by='Volume (Mn)', ascending=False)

# Select top 10 banks/payment apps
top_10 = sorted_data.head(10)

# Calculate percentage of transaction volume for each bank/payment app
top_10['Volume (%)'] = (top_10['Volume (Mn)'] / total_volume) * 100

# Visualization of transaction volume by top 10 banks/payment apps with percentages
plt.figure(figsize=(10, 6))
sns.barplot(x='UPI Banks', y='Volume (Mn)', data=top_10)
plt.title('Top 10 Banks/Payment Apps by Transaction Volume')
plt.xticks(rotation=90)
# Add percentage labels
for index, value in enumerate(top_10['Volume (%)']):
    plt.text(index, top_10['Volume (Mn)'].iloc[index] + 0.5, f'{value:.2f}%', ha='center')
plt.show()
```

# Conclusion here we can see according 2022 market key player is PhonePe app 44.18 % market share in 2021.2nd highest is Google Pay  which is having market share 34.21 in 2021

# Analyze the competitive landscape and identify the positioning of PhonePe relative to other payment 
platforms

# This analysis on 2021 Data

```python

# Group data by payment platforms and calculate relevant metrics
platform_data = df.groupby('UPI Banks').agg({
    'Volume (Mn)': 'sum',               # Total transaction volume
    'Value (Cr)': 'sum'                 # Total transaction value
}).reset_index()

# Sort data by total transaction volume in descending order
platform_data = platform_data.sort_values(by='Volume (Mn)', ascending=False)

# Select top 10 payment platforms
top_10_platforms = platform_data.head(10)

# Visualize total transaction volume by top 10 payment platforms
plt.figure(figsize=(10, 6))
sns.barplot(x='UPI Banks', y='Volume (Mn)', data=top_10_platforms)
plt.title('Total Transaction Volume by Top 10 Payment Platforms')
plt.xlabel('Payment Platforms')
plt.ylabel('Total Transaction Volume (Mn)')
plt.xticks(rotation=90)
plt.show()


```

```python
# Compare PhonePe with other payment platforms among top 10
phonepe_data = top_10_platforms[top_10_platforms['UPI Banks'] == 'PhonePe']
other_platforms_data = top_10_platforms[top_10_platforms['UPI Banks'] != 'PhonePe']

# Visualize comparison of total transaction volume between PhonePe and other platforms among top 10
plt.figure(figsize=(8, 6))
plt.bar('PhonePe', phonepe_data['Volume (Mn)'], color='blue', label='PhonePe')
plt.bar(other_platforms_data['UPI Banks'], other_platforms_data['Volume (Mn)'], color='grey', label='Other Platforms')
plt.title('Comparison of Total Transaction Volume (Top 10 Platforms)')
plt.xlabel('Payment Platforms')
plt.ylabel('Total Transaction Volume (Mn)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
