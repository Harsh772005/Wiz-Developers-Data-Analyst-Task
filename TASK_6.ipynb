```python
columns_to_drop = ['Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11', 'Column12', 'Column13', 'Unnamed: 12']
data.drop(columns=columns_to_drop, inplace=True)
#Why i deleted these columns is beacuse we are intrested to find Market analysis and these are just conversion of each other or it has Null values
```

```python
data['DATE'] = pd.to_datetime(data['DATE'], format='%d/%m/%Y')

# Extract year from the date
data['Year'] = data['DATE'].dt.year

# Step 3: Aggregate transaction volumes year-wise for PhonePe
phonepe_data = data[data['UPI APP'] == 'PhonePe']
yearly_phonepe_volumes = phonepe_data.groupby('Year')['VOL(Million)'].sum()

# Step 4: Plot year vs volume for PhonePe
plt.figure(figsize=(10, 6))
plt.plot(yearly_phonepe_volumes.index, yearly_phonepe_volumes.values, marker='o', linestyle='-')
plt.title("PhonePe's Transaction Volumes Over Time")
plt.xlabel("Year")
plt.ylabel("Transaction Volumes (Million)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data['VOL(Million)'] = pd.to_numeric(data['VOL(Million)'], errors='coerce')

# Calculate market share
phonepe_volume = data.loc[data['UPI APP'] == 'PhonePe', 'VOL(Million)'].sum()
total_volume = data['VOL(Million)'].sum()
phonepe_market_share = (phonepe_volume / total_volume) * 100
print("PhonePe's market share: {:.2f}%".format(phonepe_market_share))
```

```python
#Time Series Analysis
# Time series analysis on transaction volumes
# For example, plot monthly transaction volumes over time
monthly_transaction_volumes = data.resample('M', on='DATE')['VOL(Million)'].sum()

plt.figure(figsize=(10, 6))
plt.plot(monthly_transaction_volumes.index, monthly_transaction_volumes.values, marker='o', linestyle='-')
plt.title("Monthly Transaction Volumes Over Time")
plt.xlabel("Date")
plt.ylabel("Transaction Volumes (Million)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
